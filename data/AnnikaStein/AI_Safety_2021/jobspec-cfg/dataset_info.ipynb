```python
'''
    Derive weights for imbalanced classes during evaluation
'''
# https://towardsdatascience.com/address-class-imbalance-easily-with-pytorch-bb540497d2a6
# https://stackoverflow.com/questions/60812032/using-weightedrandomsampler-in-pytorch
test_counts = [1447281, 693971, 3139794, 18579170]

unnormalized_weight_vector = [sum(test_counts)/test_counts[i] for i in range(4)]
print(unnormalized_weight_vector)
weighted_sum_of_weights = sum([unnormalized_weight_vector[i]*test_counts[i] for i in range(4)])
print(weighted_sum_of_weights)
normalized_weight_vector = [unnormalized_weight_vector[i]/weighted_sum_of_weights for i in range(4)]
print(normalized_weight_vector)

weights = [unnormalized_weight_vector[tetarget[i]] for i in range(sum(test_counts))]
```
